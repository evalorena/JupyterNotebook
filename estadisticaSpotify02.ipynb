```python
%reset
#importar librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#leo mis datos del archivo Spotify
spotify=pd.read_csv('Spotify.csv')

#generacion de pandita (tomando una sola columna)
danza=spotify['danceability']    #OJO!!! ME EQUIVOQUE DOS VECES EN ESTE ERROR, REPASARLO

##pregunta 1
##parametro de restriccion
D_param=0.5
#num de canciones bailables
C_B=sum(danza>D_param)
#count para sacar el total
Canciones=danza.count()
#porcentaje
porcentaje = (C_B/Canciones)*100
percent_format= round(porcentaje,4)
#lo envio  pantalla
print("{}son canciones bailables de un total de {} canciones".format(str(C_B),str(Canciones)))
print("El {}% de las canciones son bailables segun un parámetro Bailable =0.5".format(percent_format))

#grafica de bailable
#metdo usando Matplotli.pyplot conocido como plt
plt.hist(danza,30)
plt.xlabel('Bailabilidad [ua]')
plt.ylabel('Número de canciones')
plt.title('Numero de canciones respecto a la bailabilidad')
plt.show()   #con el show escondo los datos del array y solo me muestra la grafica
```

```python
#Paso 3: filtra tus datos!! que queden bonitos, poblados, con columnas en enteros y lo mas pulidos posible.  
spot_filtr=spotify.filter(['acousticness','danceability','duration_ms',
                           'energy','instrumentalness','liveness', 'loudness',
                           'speechiness','tempo','time_signature','valence','song_title','artist'],axis=1)
```

```python
#paso 4: elige 2 valores/fetures/campos/variables que quieras comparar para crear tu modelo
#Nota: no debe haber datos nulos
#en este caso, tomaré el campo liveness y valence para intentar predecir cual cancion seguirá, 
#o cual cancion el usuario escuchará a continuación basados en como lo hace sentir, para detalles de valence
#ver la liga https://community.spotify.com/t5/Content-Questions/Valence-as-a-measure-of-happiness/td-p/4385221
spot_filtr_live_val=spotify.filter(['liveness','valence','song_title','artist'],axis=1)
```

```python
spot_filtr_live_val.sort_values(by='valence', ascending=False).head()
```

```python
#paso 4: elige 2 valores/fetures/campos/variables que quieras comparar para crear tu modelo
#Nota: no debe haber datos nulos
spot_live_val_dataset=spot_filtr_live_val.filter(['liveness','valence'],axis=1)
```

```python
#ahora si tomaremos el train y el test, fijamos la semilla de mis datos
#PRIMERO sacamos el tamanio de muestra
numitems = len(spot_live_val_dataset) 
print(numitems)
# fijemos 70% como entrenamiento
percenttrain = 0.7     
numtrain = int(numitems*percenttrain) #casteo a entero
numtest = numitems - numtrain 
#saco la semilla de donde voy a empezar mi toma aleatoria de datos
np.random.seed(0)
print ('Conjunto de Training con ', numtrain, 'observaciones')
print ('Conjunto de test con ', numtest, 'registros')
#creo mis dos datasets train y test
spotify_Train = spot_live_val_dataset[0:numtrain]
spotify_Test = spot_live_val_dataset[numtrain:]
```

```python
#en caso de que existan valores nulos en mis registros/ocurrencias, los elimino
spotify_Train = spotify_Train.dropna()
spotify_Test = spotify_Test.dropna()

y_train =  spotify_Train['valence']# variables de interes
y_test  = spotify_Test['valence']
del spotify_Train['valence']
del spotify_Test['valence']
spotify_Test.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier #carguemos lo minimo y suficiente
for i in range(len(spotify_Train)):
    neighbors = i+1 # experimenta cambiando el unico parametro de knn
    classifier = KNeighborsClassifier(neighbors) # inicializamos el clasificador
    classifier.fit(spotify_Train, y_train) # una sola linea hace todos los calculos :D
    predictions = classifier.predict(spotify_Test) # otra linea hace el ajuste sobre el test
    prediccionesCorrectas=[predictions == y_test] # contemos en cuantas observaciones realizamos la prediccion
    accuracy= np.sum(prediccionesCorrectas) / len(predictions)         # el calculo del error es una suma y una division :D
    print('Precisión de knn: ' + str(accuracy) + ' con ' + str(neighbors)+  ' vecinos')
```
